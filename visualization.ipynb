```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

### Line Plot (Matplotlib)

```python
x = [1,2,3,4,5]
y = [5,6,7,8,8]
plt.figure(figsize=(10,6))
plt.plot(x,y, color="orange", label ="vcxorange",linestyle="dotted",linewidth=1,marker="o",markersize=10,markeredgecolor='blue')
plt.plot(y,x, color="red", label ="line_red")
plt.ylim(1,9)
plt.xlabel("number")
plt.yticks([0,2,4,6,8])
plt.ylabel("number")
plt.title("title")
plt.grid()
plt.legend()
plt.show()
```

### Copy by Value and Reference

```python
b=5
a=b # copied by value
b=3

print(a,b)
```

```python
l1 = [1,2,3]
l2 = l1 # copied by reference
l1.append(4)

print(l1)
print(l2)
```

```python
l1 = [1,2,3]
l2 = l1.copy() # copied by value
l1.append(4)

print(l1)
print(l2)
```
