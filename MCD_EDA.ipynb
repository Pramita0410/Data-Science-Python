# Have calorie management goals?
### Let's plot a graph to check the calories per gram for different categories


```python
#practice to covert oz into g
x = "4.8 oz (136 g)"

int(x.split("(")[1].rstrip(" g)"))
```

```python
def handle_conversion(measurement):
    if "oz" in measurement and "g" in measurement:
        return int(measurement.split("(")[1].rstrip(" g)"))
    elif "fl oz cup" in measurement:
        return int(measurement.split(" ")[0])*30.46
    elif "carton" in measurement:
        return int(measurement.split("(")[1].rstrip(" ml)"))*1.03
```

```python
df["serving_size_in_grams"] = df["Serving Size"].apply(lambda x : handle_conversion(x))
```

```python
df
```

```python
df["Calories_per_gram"] = np.round(df["Calories"] / df["serving_size_in_grams"],3)
df
```

```python
plt.figure(figsize=(15,8))

sns.kdeplot(data= df, x="Calories_per_gram", hue="Category")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Calories_per_gram', data=df)
plt.title('Comparison of Caloric Density by Category')
plt.xlabel('Food Category')
plt.ylabel('Calories per Gram')
plt.xticks(rotation=70)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Calories', data=df)
plt.title('Comparison of Caloric Density by Category')
plt.xlabel('Food Category')
plt.ylabel('Calories per Gram')
plt.xticks(rotation=70)
plt.show()
```

The above graph illustrates that  'Breakfast' has the highest peak in caloric density, whereas 'Beverages' has the lowest
