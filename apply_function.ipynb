```python
data['fn'] = data['EmployeeName'].apply(lambda x: x.split(" ")[0])
data.head()
```

```python
data['Birthdate'] = pd.to_datetime(data['Birthdate'],format='%d/%m/%y')
data.info()
```

```python
from datetime import date, datetime
pd.to_datetime('today')
```

```python
data['age'] = data['Birthdate'].apply(lambda x: (pd.to_datetime('today') - x).days//365)
data.head()
```

```python
# Calculate BMI weight /height(m)**2

data['Height'] = data['Height'] / 100
data.head()
```

```python
data['bmi'] = data.apply(lambda x: x["Weight"] / x["Height"]**2, axis=1)
data.head()
```

```python
def is_healthy(bmi):
    if 25 <= bmi <= 32:
        return True
    return False
```

```python
data['is_healthy'] = data['bmi'].apply(lambda x: is_healthy(x))
data.head()
```
