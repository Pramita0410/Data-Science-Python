#### Transform

Transform merges the grouped df with the target df

```python
df.groupby('day')['tip'].mean()
```

```python
df.groupby('day')['tip'].transform('mean')
```

```python
df['mean_tip'] = df.groupby('day')['tip'].transform('mean')
df
```

#### lambda function on grp by

```python
# z score on enitre dataset

df['zscore'] = (df['total_bill'] - df['total_bill'].mean())/df['total_bill'].std()
df
```

```python
df["zscore_apply"] = df["total_bill"].apply(lambda x: (x- df["total_bill"].mean())/ df["total_bill"].std())
df
```

```python
df["zscore_wrt_dayGroup"] = df.groupby('day')["total_bill"].transform(lambda x: (x - x.mean())/x.std())
df
```

#### Apply()

```python
df.groupby("day").apply(lambda x : (x.tip/ x.total_bill).mean())
```
